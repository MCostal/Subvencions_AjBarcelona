#### GRÁFICO DE BARRAS

```python
plt.figure(figsize=(17, 12))
g = sns.histplot(data = subvencions, y = 'Entitat_Municipal', color= '#FF69B4', edgecolor = 'white')
g.bar_label(g.containers[0], padding=3, fontsize= 12)
g.set_title('Subvencions otorgades per Entitat Municipal', fontsize=15)
plt.xlabel(None)
plt.ylabel('Entitat Municipal', fontsize=13)
plt.show()
```

En el gráfico de barras, podemos ver las diferentes entidades que otorgan subvenciones con sl número total de subvenciones otorgadas. Podemos observar que el organismo que adjudica mayor número de subvenciones es el Ajuntament de Barcelona, con un total de 46280. Seguido del Institut de Cultura de Catalunya con un total de 6528. En tercer lugar, se presenta el Institut Municipal de Barcelona d'Esports con un total de 3600.

```python
seleccio = subvencions[['Entitat_Municipal','Data_Convocatoria','Data_Atorgament']]
seleccio
```

Eliminamos los registros que no tienen valor para hacer el cálculo, y calculamos la diferencia en Data Atorgament y Data Convocatoria.

```python
seleccio2 = seleccio.dropna()
seleccio2['dies'] = seleccio['Data_Atorgament'] - seleccio['Data_Convocatoria'];
```

```python
seleccio2
```

```python
seleccio2['dies'] = seleccio2['dies'].dt.days
seleccio2
```

```python
seleccio_mean = round(seleccio2.groupby('Entitat_Municipal')['dies'].mean(),1)
seleccio_mean.reset_index()
seleccio_mean = pd.DataFrame(seleccio_mean)
seleccio_mean
```

```python
plt.figure(figsize=(14, 7))
f = sns.barplot(data = seleccio_mean, y = 'Entitat_Municipal', x='dies', color = '#FF69B4', ci = 95)
f.bar_label(f.containers[0], fontsize= 12, padding = 3)
f.set_title('Dies en adjudicar una subvencio', fontsize=12)
plt.xlabel(None)
plt.ylabel('Entitat Municipal', fontsize=12)
plt.show()
```

En el gráfico de barras superior, podemos ver el tiempo medio que tarda en aprobarse una subvención desde su convocatoria por cada entidad municipal.

```python
plt.figure(figsize=(14, 7))
f = sns.boxplot(data = seleccio2, x = 'Entitat_Municipal', y='dies', color = '#FF69B4')
f.set_title('Dies en adjudicar una subvencio', fontsize=12)
plt.xlabel(None)
plt.ylabel('Entitat Municipal', fontsize=12)
plt.xticks(rotation = 45)
plt.show();
```
